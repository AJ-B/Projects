```python
#the standard
import numpy as np
import pandas as pd

#will be using synonyms of words in the emoji descriptions
from PyDictionary import PyDictionary
dictionary=PyDictionary()

#will clean the emoji descriptions using stop words
from nltk.corpus import stopwords
StopWords = stopwords.words("english")

#will be doing some punctuation cleaning
import string
```

```python
#Load emoji data
emoji_key = pd.read_csv('emojis/emDict_edited.csv', sep=';')
print emoji_key.head(5)
```

```python
#Clean the emoji descriptions for easy use later
def Clean(text):
    #REMOVE STOPWORDS
    text = ' '.join([word.lower() for word in text.split() if word not in StopWords])
    #remove punctuation
    text = text.translate(string.maketrans("",""), string.punctuation)
    return text

# EXAMPLE OF PUNCTUATION REMOVAL
# import string
# s = "string. With. Punctuation?" # Sample string 
# s.translate(string.maketrans("",""), string.punctuation)

emoji_key['Description'] = emoji_key['Description'].apply(Clean)
print emoji_key[['Description','Native']].head(15)
```

```python
#create dictionary of synonyms for each word in every description
def findSynonyms(descrip):
    descrip = descrip.split(" ")
    SynonymList = []
    for w in descrip:
        try:
            synonyms = dictionary.synonym(w)[:2]
        except:
            synonyms = []
            
        SynonymList = SynonymList + [w] + synonyms
    return " ".join(SynonymList)

emoji_key['Synonyms'] = emoji_key['Description'].apply(findSynonyms)
print emoji_key[['Description','Native','Synonyms']].head(10)
```

```python
#create dictionary of definition words for every synonym word used
def findMeaning(synonyms):
    synonyms = synonyms.split(" ")
    DefWords = []
    for s in synonyms:
        print s
        if len(s.split(" "))==1:
            try:
                definition = (dictionary.meaning(s))
                if definition==None:
                    definition=[]
                    continue
            except:
                definition = []
                d = []
            for key in definition:
                d = definition[key]
                d = " ".join(d)
                d = Clean(d).split(" ")
                DefWords = DefWords + d
    return " ".join(DefWords)

# test = ["aerial", "flying", "airy", "atmospheric", "tramway"]
# x = findMeaning(test)
# print x

emoji_key['DefWords'] = emoji_key['Synonyms'].apply(findMeaning)
print emoji_key[['Description','Native','Synonyms','DefWords']].head(5)
```

```python
#create dictionary of definition words for every synonym word used
def findMeaningDesc(descrip):
    descrip = descrip.split(" ")
    DefWords = []
    for w in descrip:
        print w
        try:
            definition = dictionary.meaning(w)
            if definition==None:
                continue
        except:
            definition = []
        d=[]  
        for key in definition:
            d = definition[key]
            d = " ".join(d)
            d = Clean(d).split(" ")
            DefWords = DefWords + d
    return " ".join(DefWords)

# test = "dango"
# x = findMeaningDesc(test)
# print "x:",type(x)

emoji_key['DescDef'] = emoji_key['Description'].apply(findMeaningDesc)
print emoji_key[['Description','Native','Synonyms','DescDef']].head(5)
```
